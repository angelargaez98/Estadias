```python
import requests
wsj= requests.get("https://www.wsj.com/")
```

```python
wsj
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(wsj.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
na = soup.find_all(class_="WSJTheme--headline--unZqjb45 reset WSJTheme--heading-3--2z_phq5h typography--serif-display--ZXeuhS5E")
```

```python
na = soup.find_all(style_="vertical-align")
na
```

```python
periods = [pt.get_text() for pt in na]
```

```python
import pandas as pd
wsj2= pd.DataFrame({
    "noticias": periods,
   
})
wsj2
```

```python
wsj2.to_csv('4-The Wall Street Journal.csv')
```
